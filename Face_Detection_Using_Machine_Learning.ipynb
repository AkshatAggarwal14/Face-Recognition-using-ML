# Face Detection


```
import numpy as np
import cv2    #(opencv)
import matplotlib.pyplot as plt
```

```
plt.imread("/content/group_photo.jpg") 
```

```
#numbers are pixel values
#can vary from 0-255
#0-very dark 255- very bright(FOR BW IMAGES)
#size-7X7(LET 7)
#FOR RGB VALUE there will be 3 matrixes behind each other
#each matrix having pixel value map for each color
#size -7X7X3
#or ./group_photo.jpg here ./ means in this folder
img=plt.imread("/content/group_photo.jpg") 

```

```
img.shape
```

```
plt.imshow(img)
```

```
plt.imshow(img[:,:,0])
```

```
red_img=img.copy()
red_img[: ,: ,[1,2]]=0
plt.imshow(red_img)#[1,2] for red [0,1] blue [0,2] green
```

```
#OPENCV has trained ML algo with faces so easy recognize faces
#Open cv haarcascade files google
#haarcascade_frontalface_default.xml <-- used
```

```
model=cv2.CascadeClassifier("./haarcascade_frontalface_default.xml")
```

```
model.detectMultiScale(img) #<--finds faces
#this return 6x4 2D Matrix 6=Number of faces and 4=number of properties
#properties are(X coord,Y coord, width, height)
#x coord and y coord are coords of top left corner
```

```
faces=model.detectMultiScale(img)
```

```
faces[1]
```

```
x,y,w,h=faces[1]
```

```
print(x)
print(y)
print(w)
print(h)
```

```
cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
#here (x,y) are pt1 coords, other is pt2 coords
#(255,0,0) represents red colour
#2 is thickness iof rectangle drawn n pixels
#this returns updated image data
```

```
plt.imshow(img)
```

```
for ele in faces:
  print(ele)
```

```
for ele in faces:
  x,y,w,h=ele
  img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2) #updating image for every face
```

```
plt.imshow(img)
```
